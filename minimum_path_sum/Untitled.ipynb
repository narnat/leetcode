```python
from pandas import *
import matplotlib as mpl
import matplotlib.pyplot as plt
plt.style.use('classic')
```

```python
class Solution:
    def minPathSum(self, grid):
        if not grid:
            return 0
        
        print("Initial stage\n")
        print(DataFrame(grid), "\n")
        
        col = len(grid)
        row = len(grid[0])
        
        for i in range(1, row):
            grid[0][i] += grid[0][i - 1]

        print("After 1st loop\n")
        print(DataFrame(grid), "\n")
        for i in range(1, col):
            
            grid[i][0] += grid[i - 1][0]
            print("After sum of col", i)
            print(DataFrame(grid), "\n")
            
            for j in range(1, row):
                grid[i][j] += min(grid[i - 1][j], grid[i][j - 1])
                print("Calling min for i and j", i, j)
                print(DataFrame(grid), "\n")
                print("****************************************")
        
        print("Last stage\n")
        print(DataFrame(grid), "\n")
        return grid[col - 1][row  - 1]
```

```python
minPathSum = Solution().minPathSum

grid = [[1,3,1],[1,5,1],[4,2,1]]
print(minPathSum(grid))
print("======================================================================")
grid = [[1,2],[1,1]]
print(minPathSum(grid))
```
